```python
import pandas as pd
```

```python
Counties_Vote = pd.read_csv('/content/drive/My Drive/Datasets/Project_2/Voting-Counties.csv')
```

```python
df = pd.DataFrame(Counties_Vote)
```

```python
df.tail(10)
```

#Task 1

```python
Total_Votes_Per_Candidate = df.groupby(['state_name'])[['Votes', 'BB Votes', 'RR Votes']].sum()
```

```python
print(Total_Votes_Per_Candidate)
```

```python
Total_Votes_Per_Candidate['Total'] = Total_Votes_Per_Candidate['BB Votes'] + Total_Votes_Per_Candidate['RR Votes']
```

```python
print(Total_Votes_Per_Candidate['Total'])
```

#Task 2

```python
Total_Votes_Per_Candidate['BB_Percentage'] = (Total_Votes_Per_Candidate['BB Votes'] / Total_Votes_Per_Candidate['Total']) * 100
Total_Votes_Per_Candidate['RR_Percentage'] = (Total_Votes_Per_Candidate['RR Votes'] / Total_Votes_Per_Candidate['Total']) * 100
```

```python
print(Total_Votes_Per_Candidate[['BB_Percentage', 'RR_Percentage']])
```

```python
Total_Votes_Per_Candidate
```

#Task 3

```python
total_rr_votes = df['RR Votes'].sum()
total_bb_votes = df['BB Votes'].sum()
```

```python
popular_vote_winner = "Road Runner" if total_rr_votes > total_bb_votes else "Buggs Bunny"
print(f'The candidate with the popular vote is {popular_vote_winner} with Road Runner Votes at {total_rr_votes} and Bug Bunny votes at {total_bb_votes}')
```

#Task 4

```python
State_Info = pd.read_csv('/content/drive/MyDrive/Datasets/Project_2/State-Info.csv')
```

```python
df1 = Total_Votes_Per_Candidate.merge(State_Info[['Location', 'Electoral Votes']], left_on='state_name', right_on='Location', how='left')
```

```python
df1.head(10)
```

#Task 5

```python
df1['State_Winner'] = df1.apply(lambda x: 'BB' if x['BB Votes'] > x['RR Votes'] else 'RR', axis=1)
```

```python
df1['Electoral_Votes_Won_BB'] = df1.apply(lambda x: x['Electoral Votes'] if x['State_Winner'] == 'BB' else 0, axis=1)
df1['Electoral_Votes_Won_RR'] = df1.apply(lambda x: x['Electoral Votes'] if x['State_Winner'] == 'RR' else 0, axis=1)
```

```python
df1.head(10)
```

#Task 6

```python
total_electoral_votes_bb = df1['Electoral_Votes_Won_BB'].sum()
total_electoral_votes_rr = df1['Electoral_Votes_Won_RR'].sum()
```

```python
print("Total Electoral Votes for BB:", total_electoral_votes_bb)
print("Total Electoral Votes for RR:", total_electoral_votes_rr)
```
