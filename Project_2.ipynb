```python
import pandas as pd
```

```python
Counties_Vote = pd.read_csv('/content/drive/My Drive/Datasets/Project_2/Voting-Counties.csv')
```

```python
df = pd.DataFrame(Counties_Vote)
```

```python
df.head(10)
```

#Task 1 (Collect Votes cast from counties in the County_Votes.csv and determine the sum of votes for candidate in each state)

```python
#Grouping all the counties under their state and summing up the total votes, BB votes, and RR votes for each state

Total_Votes_Per_Candidate = df.groupby(['state_name'])[['Votes', 'BB Votes', 'RR Votes']].sum()
```

```python
print(Total_Votes_Per_Candidate)
```

```python
#Creating the column, Total, in the newly created dataset

Total_Votes_Per_Candidate['Total'] = Total_Votes_Per_Candidate['BB Votes'] + Total_Votes_Per_Candidate['RR Votes']
```

```python
print(Total_Votes_Per_Candidate['Total'])
```

#Task 2 (Convert the votes to % win for Buggs Bunny and for Road Runner in each state.)

```python
#Calculating the percentage by diving the candidates votes by the total amount of votes. After that, Multiplying it by 100 to get it into percentage form

Total_Votes_Per_Candidate['BB_Percentage'] = (Total_Votes_Per_Candidate['BB Votes'] / Total_Votes_Per_Candidate['Total']) * 100
Total_Votes_Per_Candidate['RR_Percentage'] = (Total_Votes_Per_Candidate['RR Votes'] / Total_Votes_Per_Candidate['Total']) * 100
```

```python
print(Total_Votes_Per_Candidate[['BB_Percentage', 'RR_Percentage']])
```

```python
Total_Votes_Per_Candidate
```

#Task 3 (Determine who won the popular vote)

```python
#Getting the total amount of votes for each candidate

total_rr_votes = df['RR Votes'].sum()
total_bb_votes = df['BB Votes'].sum()
```

```python
#Using ternary operator to check who won the popular vote

popular_vote_winner = "Road Runner" if total_rr_votes > total_bb_votes else "Buggs Bunny"
print(f'The candidate with the popular vote is {popular_vote_winner} with Road Runner Votes at {total_rr_votes} and Bug Bunny votes at {total_bb_votes}')
```

#Task 4 (Read in the info from State-Info.csv)

```python
#Reading in the State Info file to get the electoral votes

State_Info = pd.read_csv('/content/drive/MyDrive/Datasets/Project_2/State-Info.csv')
```

```python
#Merging the two created datasets so that the computer doesn't get confused
#Better and easier usage if it was within the same dataset

df1 = Total_Votes_Per_Candidate.merge(State_Info[['Location', 'Electoral Votes']], left_on='state_name', right_on='Location', how='left')
```

```python
#Printing df1 to see if everything merged over

df1.head(10)
```

#Task 5 (Award Electoral Votes for each state to the winner of the state with the highest number of votes for either BB or RR)

```python
#Creating the column, State Winner, to go through each column and see who won each state
#Lambda is used as we will be only using the function once

df1['State_Winner'] = df1.apply(lambda x: 'BB' if x['BB Votes'] > x['RR Votes'] else 'RR', axis=1)
```

```python
#Creating the two electoral votes column and awards them the electoral votes

df1['Electoral_Votes_Won_BB'] = df1.apply(lambda x: x['Electoral Votes'] if x['State_Winner'] == 'BB' else 0, axis=1)
df1['Electoral_Votes_Won_RR'] = df1.apply(lambda x: x['Electoral Votes'] if x['State_Winner'] == 'RR' else 0, axis=1)
```

```python
df1.head(10)
```

#Task 6 (Determine who won the Electoral Vote tally)

```python
#Summing the Electoral Votes for each candidate to see who won the race

total_electoral_votes_bb = df1['Electoral_Votes_Won_BB'].sum()
total_electoral_votes_rr = df1['Electoral_Votes_Won_RR'].sum()
```

```python
print("Total Electoral Votes for BB:", total_electoral_votes_bb)
print("Total Electoral Votes for RR:", total_electoral_votes_rr)
```

#Task 7 (Accept the State ID and list all of the summary info for the state with winner, totals and electoral votes awarded.)

```python
#Adding in the State IDs

df1["State_id"] = State_Info['Postal Abbreviation']
```

```python
#Reindexing df1 to make it look more organized and easier to read each data column

df1 = df1.reindex(columns=['Location', 'State_id', 'Votes', 'BB Votes', 'RR Votes', 'BB_Percentage', 'RR_Percentage', 'Electoral Votes', 'State_Winner', 'Electoral_Votes_Won_BB', 'Electoral_Votes_Won_RR'])
```
